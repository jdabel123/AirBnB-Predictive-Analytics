## 3. Data Preparation

- Filling missing values with the mean value for the respective column.
- Handling categorical data by creating dummy variables.

```python
#Cleaning data

"""
I chose to impute the missing with the mean for the respective columns. I chose this method instead of dropping
the rows with missing values because the largest percentage of missing values was 29% for cleaning_fee.

"""

df['host_response_rate'] = df['host_response_rate'].str.replace('%','').astype('float64')
df['host_acceptance_rate'] = df['host_acceptance_rate'].str.replace('%','').astype('float64')

df['host_is_superhost'] = np.where(df['host_is_superhost'] == 't',1,0)

df['price'] = df['price'].str.replace('[$, ]','').astype('float64')

df['cleaning_fee'] = df['cleaning_fee'].str.replace('[$, ]','').astype('float64')
    
```

```python
#Filling in missing values
fill_missing = lambda col:col.fillna(col.mean())    
    
def filling_numerical_missing_values(df,function):
    """
    Function to fill in the missing values in the dataframe that are numerical data types with the desired value
    
    INPUT:
    
    df - The dateframe that you want to impute missing values
    
    function - The desired function to impute missing values
    
    OUTPUT:
    
    The function will return the data frame with imputed missing values for numerical data types.
    """
    
    
    num_vars = df.select_dtypes(include=['float64','int64']).copy()
    
    filled_num_vars = num_vars.apply(function,axis=1)
    
    # Reassigning columns in original data frame with the columns in filled_num_vars
    
    for col in filled_num_vars:
        df[col] = filled_num_vars[col]
        
    return df

df = filling_numerical_missing_values(df,fill_missing)

```

## Creating a Linear Regression model to try and predict price.


Creating dummy variables for categorical data


```python
sns.heatmap(df.corr(),cbar=True);
```

```python
#Getting categorical data

cat_var = df.select_dtypes(include='object')

cat_var
```

```python
for col in cat_var.columns:
    df = pd.concat([df.drop(col,axis=1), pd.get_dummies(df[col], prefix=col, prefix_sep='-', drop_first=True)],axis=1)
```

```python
df
```

## 4. Modelling

- Using a linear regression model to try and predict the listing price based on a variety of variables.

```python
#Split the data into test and train set 

X = df.drop('price',axis=1)

y = df['price']


X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.3, random_state = 42)

#Instantiate model: I have chosen fit_intercept = False because when fit_intercept = False the model predicts very high values
#for the test data.

lm_model = LinearRegression(normalize=True,fit_intercept=False)


#Train the model

lm_model.fit(X_train,y_train)

#Predict using x test values

y_test_pred = lm_model.predict(X_test)

y_train_pred = lm_model.predict(X_train)


#Measure r2 score

test_r2_score = r2_score(y_test,y_test_pred)
train_r2_score = r2_score(y_train,y_train_pred)

print("Test R2 score: {}".format(test_r2_score))
print("Train R2 score: {}".format(train_r2_score))

```

## 5. Evaluation

```python
plt.scatter(y_train_pred,y_train_pred - y_train,color = 'dodgerblue')
plt.scatter(y_test_pred,y_test_pred - y_test, color='crimson',alpha=.6)

plt.ylabel('Residuals')
plt.xlabel('Price')

plt.title("The residuals from Train and Test values")

plt.legend(['Training data','Test data'],frameon=False);
```

```python
coefficients = pd.concat([pd.DataFrame(X.columns),pd.DataFrame(np.transpose(lm_model.coef_))], axis = 1)

column_names = ['Features','coef_']

coefficients.columns = column_names

coefficients = coefficients.sort_values(by='coef_',ascending=False).reset_index()

coefficients.head()
```
